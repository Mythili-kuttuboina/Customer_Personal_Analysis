```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# 1. Selecting the most relevant features for clustering
# We include spending, income, and engagement metrics
features = ['Income', 'MntWines', 'MntMeatProducts', 'MntGoldProds', 
            'NumWebPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']

# Drop rows with missing values in these specific columns
data_ml = data.dropna(subset=features)

# 2. Standardizing the data (Crucial for K-Means)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(data_ml[features])
```

```python
inertia = []
range_val = range(1, 11)

for i in range_val:
    k_means = KMeans(n_clusters=i, random_state=42)
    k_means.fit(scaled_features)
    inertia.append(k_means.inertia_)

plt.figure(figsize=(10,6))
plt.plot(range_val, inertia, marker='o', color='teal')
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# Applying K-Means with 4 clusters
model = KMeans(n_clusters=4, random_state=42)
data_ml['Cluster'] = model.fit_predict(scaled_features)

# Checking the size of each segment
print(data_ml['Cluster'].value_counts())
```

```python
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_features)
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x=pca_data[:,0], y=pca_data[:,1], hue=data_ml['Cluster'], palette='Set1')
plt.title('Customer Segments Visualization (PCA)')
plt.show()
```

```python
# Analyze the characteristics of each cluster
analysis = data_ml.groupby('Cluster')[features].mean()
print(analysis)
```
